```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

plt.rcParams.update({"figure.autolayout": True})

def simulasi_gerak_lurus(s0, v0, a, m, t_total, dt, g=9.8):
    times = np.arange(0, t_total + 1e-12, dt)
    s_list, v_list, ek_list, ep_list, etot_list = [], [], [], [], []
    for t in times:
        s = s0 + v0 * t + 0.5 * a * t**2
        v = v0 + a * t
        ek = 0.5 * m * v**2
        ep = m * g * s
        etot = ek + ep
        s_list.append(s)
        v_list.append(v)
        ek_list.append(ek)
        ep_list.append(ep)
        etot_list.append(etot)
    df = pd.DataFrame({
        "t (s)": times,
        "s (m)": np.round(s_list, 5),
        "v (m/s)": np.round(v_list, 5),
        "E_kinetik (J)": np.round(ek_list, 5),
        "E_potensial (J)": np.round(ep_list, 5),
        "E_total (J)": np.round(etot_list, 5)
    })
    return df

def tampilkan_hasil(df, title):
    display(df)
    plt.figure()
    plt.plot(df["t (s)"], df["s (m)"])
    plt.xlabel("t (s)"); plt.ylabel("s (m)")
    plt.title(f"{title} — Posisi vs Waktu")
    plt.grid(True); plt.show()
    plt.figure()
    plt.plot(df["t (s)"], df["v (m/s)"])
    plt.xlabel("t (s)"); plt.ylabel("v (m/s)")
    plt.title(f"{title} — Kecepatan vs Waktu")
    plt.grid(True); plt.show()
    plt.figure()
    plt.plot(df["t (s)"], df["E_kinetik (J)"])
    plt.xlabel("t (s)"); plt.ylabel("E_kinetik (J)")
    plt.title(f"{title} — Energi Kinetik vs Waktu")
    plt.grid(True); plt.show()
    plt.figure()
    plt.plot(df["t (s)"], df["E_potensial (J)"])
    plt.xlabel("t (s)"); plt.ylabel("E_potensial (J)")
    plt.title(f"{title} — Energi Potensial vs Waktu")
    plt.grid(True); plt.show()
    plt.figure()
    plt.plot(df["t (s)"], df["E_total (J)"])
    plt.xlabel("t (s)"); plt.ylabel("E_total (J)")
    plt.title(f"{title} — Energi Total vs Waktu")
    plt.grid(True); plt.show()

m, a, v0, s0, t_total, dt = 3.0, 1.5, 2.0, 12.0, 10, 2.0
df4 = simulasi_gerak_lurus(s0, v0, a, m, t_total, dt)
print("=== Kasus 4 ==="); tampilkan_hasil(df4, "Kasus 4"); df4.to_csv("kasus4.csv", index=False)
```
