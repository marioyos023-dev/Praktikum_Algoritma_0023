```python
import matplotlib.pyplot as plt
v0 = float(input("Masukkan kecepatan awal (m/s): "))
a = float(input("Masukkan percepatan (m/s^2): "))
t_total = int(input("Masukkan waktu total (detik): "))

waktu = []
posisi = []

for t in range(t_total + 1):
    s = v0 * t + 0.5 * a * (t**2)
    waktu.append(t)
    posisi.append(s)

if a > 0:
  print("Benda mengalami percepatan.")
elif a < 0:
  print("Benda mengalami perlambatan.")
else:
  print("Benda bergerak dengan kecepatan konstan.")

plt.figure()
plt.plot(waktu, posisi, marker='o')
plt.title("Grafik Posisi terhadap waktu pada gerak lurus berubah beraturan")
plt.xlabel("Waktu (detik)")
plt.ylabel("Posisi (meter)")
plt.grid(True)
plt.show()
```
