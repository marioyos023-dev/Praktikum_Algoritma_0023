```python
# ============================================
# PROGRAM OPERASI MATRIKS – 4 KASUS (MODUL 5)
# ============================================

def input_matriks(baris, kolom, nama):
    print(f"\nMasukkan elemen matriks {nama} ({baris}x{kolom}):")
    matriks = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            while True:
                try:
                    val = float(input(f"Elemen [{i+1},{j+1}]: "))
                    row.append(val)
                    break
                except ValueError:
                    print("Input harus angka!")
        matriks.append(row)
    return matriks


def print_matriks(M):
    for row in M:
        print("\t".join(f"{x:.2f}" for x in row))


def tambah(A, B):
    baris = len(A)
    kolom = len(A[0])
    C = []
    for i in range(baris):
        baris_baru = []
        for j in range(kolom):
            baris_baru.append(A[i][j] + B[i][j])
        C.append(baris_baru)
    return C


def kali(A, B):
    m = len(A)
    p = len(A[0])
    n = len(B[0])
    C = []
    for i in range(m):
        row = []
        for j in range(n):
            total = 0
            for k in range(p):
                total += A[i][k] * B[k][j]
            row.append(total)
        C.append(row)
    return C



def transpose(A):
    baris = len(A)
    kolom = len(A[0])
    T = []
    for j in range(kolom):
        row = []
        for i in range(baris):
            row.append(A[i][j])
        T.append(row)
    return T


# ========================
# PROGRAM UTAMA 4 KASUS
# ========================

def jalankan_kasus(baris_A, kolom_A, baris_B, kolom_B, nomor_kasus):
    print("\n===================================")
    print(f"           KASUS {nomor_kasus}")
    print("===================================")

    A = input_matriks(baris_A, kolom_A, "A")
    B = input_matriks(baris_B, kolom_B, "B")

    while True:
        print("\nMENU OPERASI:")
        print("1. Penjumlahan")
        print("2. Perkalian")
        print("3. Transpose")
        print("4. Kembali ke menu utama")

        pilih = input("Pilih menu: ")

        if pilih == "1":
            if (len(A) == len(B)) and (len(A[0]) == len(B[0])):
                print("\nHASIL A + B:")
                print_matriks(tambah(A, B))
            else:
                print("\nPenjumlahan tidak dapat dilakukan (ukuran beda).")

        elif pilih == "2":
            if len(A[0]) == len(B):
                print("\nHASIL A × B:")
                print_matriks(kali(A, B))
            else:
                print("\nPerkalian tidak bisa dilakukan (kolom A != baris B).")

        elif pilih == "3":
            print("\nTranspose Matriks A:")
            print_matriks(transpose(A))
            print("\nTranspose Matriks B:")
            print_matriks(transpose(B))

        elif pilih == "4":
            break

        else:
            print("Pilihan tidak valid!")


def main():
    while True:
        print("\n===================================")
        print("      PROGRAM 4 KASUS MATRIKS")
        print("===================================")
        print("1. Kasus 1 (A: 2x2, B: 2x2)")
        print("2. Kasus 2 (A: 2x3, B: 3x2)")
        print("3. Kasus 3 (A: 3x3, B: 3x3)")
        print("4. Kasus 4 (A: 3x2, B: 3x2)")
        print("5. Keluar Program")

        menu = input("Pilih kasus: ")

        if menu == "1":
            jalankan_kasus(2, 2, 2, 2, 1)
        elif menu == "2":
            jalankan_kasus(2, 3, 3, 2, 2)
        elif menu == "3":
            jalankan_kasus(3, 3, 3, 3, 3)
        elif menu == "4":
            jalankan_kasus(3, 2, 3, 2, 4)
        elif menu == "5":
            print("Program selesai.")
            break
        else:
            print("Pilihan tidak valid!")


main()
```
